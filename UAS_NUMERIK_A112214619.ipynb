```python
#FISIKA 1 Hukum 1 Newton

# Massa benda dalam kilogram
m = 10  # kg

# Percepatan gravitasi dalam meter per detik kuadrat
g = 10  # m/s^2

# Menghitung gaya gravitasi
Fg = m * g

# Karena percepatan benda nol, tegangan tali sama dengan gaya gravitasi
T = Fg

print(f"Besarnya tegangan tali (T) adalah: {T} N")
```

```python
#FISIKA 1 Hukum 2 Newton

# Massa benda dalam kilogram
m1 = 20  # kg
m2 = 5   # kg

# Percepatan gravitasi dalam meter per detik kuadrat
g = 10  # m/s^2

# Menghitung percepatan
a = (m2 * g) / (m1 + m2)

# Menghitung tegangan tali
T = m1 * a

print(f"Percepatan (a) adalah: {a} m/s^2")
print(f"Tegangan tali (T) adalah: {T} N")

```

```python
#FISIKA 2
# Operasi vektor: perkalian dari dua buah vektor
# Torsi didefinisikan sebagai: v = r x F.
# Bila sebuah gaya F = (6 i + 10 k) N
# dipakai untuk memutar sebuah kunci pas dan posisi baut terhadap titik
# tangkap gaya adalah r = (-4 i + 10 j) m.
# Hitung torsi yang dilakukan gaya tersebut--

import numpy as np
# Vektor gaya F
F = np.array([6, 0, 10])
# Vektor posisi r
r = np.array([-4, 10, 0])
# Perkalian silang (cross product)
torsi = np.cross(r, F)
print("Torsi yang dilakukan gaya tersebut adalah:", torsi)
```

```python
#FISIKA 2 Gauss's Law : Spherical Symmerty

import numpy as np

# Konstanta
epsilon_0 = 8.854e-12  # Permitivas vakum dalam F/m

# Parameter
Q = 1e-6  # Muatan total dalam Coulomb (misalnya 1 mikroCoulomb)
R = 0.1   # Radius bola dalam meter (misalnya 10 cm)

def electric_field(r, Q, R, epsilon_0):
    if r < R:
        # Medan listrik di dalam bola (r < R)
        E = (Q * r) / (4 * np.pi * epsilon_0 * R**3)
    else:
        # Medan listrik di luar bola (r â‰¥ R)
        E = (Q) / (4 * np.pi * epsilon_0 * r**2)
    return E

def main():
    # Jarak dari pusat bola dalam meter
    r_values = [0.05, 0.1, 0.15]  # Contoh beberapa jarak (misalnya 5 cm, 10 cm, 15 cm)

    for r in r_values:
        E = electric_field(r, Q, R, epsilon_0)
        print(f"Medan listrik pada r = {r} m adalah: {E:.2e} N/C")

if __name__ == "__main__":
    main()
```
